```python
def yolo_to_unnormalized_box(yolo_box, image_width, image_height):
    center_x, center_y, width, height = yolo_box

    # Convert normalized coordinates to original image coordinates
    x_center = int(center_x * image_width)
    y_center = int(center_y * image_height)
    box_width = int(width * image_width)
    box_height = int(height * image_height)

    # Calculate xmin, xmax, ymin, ymax
    x_min = x_center - box_width // 2
    y_min = y_center - box_height // 2
    x_max = x_min + box_width
    y_max = y_min + box_height

    return x_min, x_max, y_min, y_max
```

```python
imgfilesPath=os.path.join(dest,"Images")
npzfilesPath=os.path.join(dest,"NPZFiles")
```

```python
textfilesPath=os.path.join(dest,"YOLOTextFiles")
```

```python
npzfiles=os.listdir(npzfilesPath)
```

```python
for npzfile in npzfiles:
    maskpath=os.path.join(npzfilesPath,npzfile)
    textpath=os.path.join(textfilesPath,npzfile[:-3]+"txt")
    print("Paths Obtained:",maskpath,textpath)
    f=open(textpath,'w')
    masks=np.load(maskpath)
    masks=masks['arr_0']
    h=(masks.shape)[0]
    w=(masks.shape)[1]
    n=(masks.shape)[2]
    lines=[]
    for i in range (0,n):
        line="0"
        mask=(masks[:,:,i]).astype(np.uint8)
        contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        for j in range(0,len(contours)):
            cont=contours[j]
            for k in range(0,len(cont)):
                contour=cont[k][0]
                x=contour[0]/w
                y=contour[1]/h
                line+=" {} {}".format(x,y)
        # print(line)
        lines.append(line+'\n')
    f.writelines(lines)
    f.close()
    
```
